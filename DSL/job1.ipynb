## DSL Job1

The program will consist of 2 functions. The first one will check whether the provided string is a filename and read its data or treat the given string as a text directly. The second function will split the text into chunks based on the provided whitespaces.

Let's take a look at the first function. It accepts the `pathOrText` string as a parameter, checks if it is a filename and returns its content, otherwise returns the provided string as is.

```python
import os.path

def getText(pathOrText):
    return open(pathOrText, 'r').read() if os.path.isfile(pathOrText) else pathOrText
```

The second function splits the text into the blocks via the provided whitespaces. The flow is the following: firstly, get the actual text from the provided string by calling the `getText` method. Secondly, loop through the text symbol by symbol and check if a symbol belongs to the whitespaces set. If the symbol is a whitespace symbol, append the local result to the final result array and clear the local result, else append the read symbol to the local result string. In the end check if the local result is not empty, append it to the final result and return the final result.

```python
def split(pathOrText, whitespaces):
  text = getText(pathOrText)
    
  localResult = ""
  result = []
    
  for symbol in text:
    if symbol in whitespaces:
      if (len(localResult) > 0):
        result.append(localResult)
      localResult = ""
    else:
      localResult += symbol
    
  if (len(localResult) > 0):
    result.append(localResult)
    
  return result
```

### Test:

```python
text = "abracadabra"
whitespaces = {"a"}

assert split(text, whitespaces) == ["br", "c", "d", "br"]
```
